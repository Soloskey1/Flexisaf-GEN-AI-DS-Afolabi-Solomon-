## Module 3: Data Visualization with Matplotlib and Seaborn

In the module, we will explore the use on seaborn and matplotlib for data visualization. We will create basic plots using these python libraries. we are making use of health data which can be downloaded from kaggle.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('Heart_health.csv')
df.head(10)
```

```python
# shape of data
df.shape
```

```python
# information about the data 
df.info()
```

The data is clean with no null values, since we have a total of 724 non_null values, same as the number of total entries.

```python
# description 
df.describe().T
```

**1. Visualization of the distribution smokers and non-smokers by gender.**

```python
# group smokers and non-smokers by gender
plt.figure(figsize=(10,7))
sns.violinplot(x='Gender', y='Age', data=df, palette='Set2', hue='Smoker')
plt.title('Number of Males and Females by smoking habits')
```

**2. Proportion of males and female patientadmitted**

```python
# total number of patients
plt.figure(figsize=(9,9))
heart_attack_by_gender = df.groupby('Gender')['Heart Attack'].sum()
heart_attack_by_gender.plot(kind='pie', autopct='%1.1f%%',
                            title = 'Number of male and female patients')
```

**3. Visualize the the smoking habits of patients and their heart health status**

```python
# plot the level of cholestrol by age

#sample some data from the dataset
plt.figure(figsize=(10,10))
sampled_df = df.sample(n=250, random_state=42)
sns.boxplot(x='Age', y='Smoker', data=sampled_df, hue='Heart Attack', palette='bright')
plt.xlabel('Age')
plt.ylabel('Smoker')
plt.title('Box plot of smooking habits of patients')
```

**4. Visualization of diabetic and non-diabetic patients by their weights**

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
sns.lineplot(ax =axes[0], data=df, x="Weight(kg)", y="Glucose(mg/dL)",  hue="Heart Attack")
sns.lineplot(ax = axes[1], data=df, x="Cholesterol(mg/dL)", y="Glucose(mg/dL)",  hue="Heart Attack")
axes[0].set_title("Distribution of patient with heart attacks by Weight")
axes[1].set_title("Distribution of patient with heart attacks by Cholesterol level")
```

**5. Visualize between some features in the data**

```python
plt.figure(figsize=(9,8))
pair_plot = sns.pairplot(df, 
             x_vars = ['Cholesterol(mg/dL)', 'Glucose(mg/dL)', 'Exercise(hours/week)'],
             y_vars = ['Cholesterol(mg/dL)', 'Glucose(mg/dL)', 'Exercise(hours/week)'],
             hue='Heart Attack', palette='Set2', diag_kind='hist') 

# to adjust the layout to make room for the title
pair_plot.fig.subplots_adjust(top=0.95)

# Add a title to the figure
pair_plot.fig.suptitle('Relationship between some features', fontsize=16)
plt.show()
```

**6. Use box plot to viaulaize and identify in the cholesterol levels betweern age groups**

```python
# create buckets to hold ages between intervals
bins = [0, 20, 30, 40, 50, 60, 70, 80]
labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80']

# ceeate new column with age groups
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

plt.figure(figsize=(10, 6))
sns.boxplot(data = df, x = 'Age_Group', y = 'Cholesterol(mg/dL)', hue='Heart Attack').set_title('Box plot of Cholesterol levels within age groups')
```

**7. Visualize the distribution of age and weights by heart health status**

```python
sns.scatterplot(data=df, x = 'Age', y = 'Weight(kg)', hue='Heart Attack')
```

**8. Create a Correlation plot**

Correlation plot is used to identify the relationship between features in a given data. It displays points over box, where each box shows a point of contact between invidual feature with another, It ranges from -1 to 1. That is from strong negetive relationship to strong positive relation, 0 meaning there is no obvious between the features.

```python
cols = ['Age', 'Cholesterol(mg/dL)', 'Glucose(mg/dL)', 'Exercise(hours/week)',
        'Height(cm)', 'Weight(kg)']
sns.heatmap(data= df[cols].corr(), annot=True, fmt='.2f')
```
