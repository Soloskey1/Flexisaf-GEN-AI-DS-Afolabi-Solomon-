**1. Create a program to create a fucntion func1() to accept a variable length arguments and print their value.**

```python
# this program will accept variable number of elements and print each item ('ele')
def func1(*elements): 
    for ele in elements:
        print(ele)
```

**2. Write a program to create a recursive function to calculate the sum of number from 0 tp 15.**

```python
# I gave the function a default value of 15 since it is supposed to recursively add
# numbers from 0 t0 15
def recursive_sum(n=15):
    # if n is 0, return 0
    if n == 0:
        return 0
    else:
        # return n plus sum of numbers from n to n - 1
        return n + recursive_sum(n - 1)

n = 15
print(f'The sum of numbers from 0 to {n} is {recursive_sum(n)}')
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('company_sales_data.csv')
sales.head()
```

```python
# a) 1. 
# Total profits of all months 

# group by months
group_by_months = sales.groupby('month_number')['total_profit'].sum()
print(group_by_months)
```

The 'month-number' is the index while the 'total_profit' for each month are the values. 

```python
# Visualize using line plot 

# Line plot
plt.figure(figsize=(9,7))
plt.plot(group_by_months.index, group_by_months.values, linestyle='-', 
         color = 'b', marker='o')
# titles and labels
plt.title('Sum of Total Profit by Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
# to ensure that all month numbers are shown
plt.xticks(group_by_months.index)
plt.grid(True)
plt.show()
```

```python
# b) 2. 

# bathing soap (bs) and face wash (fw)
# to get the total sales of bathing soap of each month
bs_fw = sales.groupby('month_number')[['bathingsoap', 'facewash']].sum()
bs_fw
```

```python
#  to use pltsubplots to plot both bathingsoap and facewash

# create subplots to contain line plots for bathing soap and facewash 
# share values in x-axis
fig, axes = plt.subplots(2,1, figsize = (9,8), sharex = True)

# plotting for 'bathingsoap'
axes[0].plot(bs_fw.index, bs_fw['bathingsoap'], marker='o', color='g', label='Bathing Soap')
axes[0].set_title('Sales of Bathing Soap')
axes[0].set_ylabel('Sales')
axes[0].legend()
axes[0].grid(True)

# Plotting for 'facewash'
axes[1].plot(bs_fw.index, bs_fw['facewash'], marker='o', color='r', label='Facewash')
axes[1].set_title('Sales of Facewash')
axes[1].set_xlabel('Month NUmber')
axes[1].set_ylabel('Sales')
axes[1].legend()
# to show all moth numbers
plt.xticks(group_by_months.index)
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
